```python
# Imports
import pandas as pd
```

### Análisis de la tabla Clientes

```python
df_clientes = pd.read_excel(r"Base de datos/clientes.xlsx")
df_clientes.head(3)
```

```python
df_clientes.describe(include='all').round(2)
```

```python
df_clientes['ciudad'].value_counts(normalize=True).plot(kind='bar', title='Distribución de clientes por ciudad')
```

### Análisis de la tabla Productos

### Análisis de la tabla Ventas

```python
df_detalle_ventas = pd.read_excel(r"Base de datos/Detalle_ventas.xlsx")
df_detalle_ventas.head(3)
```

```python
df_detalle_ventas.describe(include='all').round(2)
```

```python
import matplotlib.pyplot as plt

df_productos['precio_unitario'].hist(bins=25)
plt.title('Histograma de precio_unitario')
plt.xlabel('Precio unitario')
plt.ylabel('Frecuencia')
plt.show()

print("Media:", df_productos['precio_unitario'].mean())
print("Mediana:", df_productos['precio_unitario'].median())
print("Moda:", df_productos['precio_unitario'].mode()[0])
```

```python
df_detalle_ventas['cantidad'].hist(bins=7)
plt.title('Histograma de cantidad')
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')
plt.show()

print("Media:", df_detalle_ventas['cantidad'].mean())
print("Mediana:", df_detalle_ventas['cantidad'].median())
print("Moda:", df_detalle_ventas['cantidad'].mode()[0])
```
